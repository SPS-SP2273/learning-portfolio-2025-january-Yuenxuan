```python
# Some data for plotting
x = [0, 1, 2, 3, 4, 5, 6]
y_1 = [0, 2, 4, 6, 8, 10, 12]
y_2 = [0, 3, 6, 9, 12, 15, 18]

# Let's start plotting
plt.plot(x, y_1, label='Y values',
         color='blue', linestyle='dashed')
plt.xlabel('time')
plt.ylabel('depression')
plt.title('X vs Y')
plt.grid(alpha=.25)
plt.legend(loc='upper left')
#plt.show()
#plt.savefig("XvsY.jpg") png pdf can also 
```

```python
# Some data for plotting
x = [0, 1, 2, 3, 4, 5, 6]
y_1 = [0, 2, 4, 6, 8, 10, 12]
y_2 = [0, 3, 6, 9, 12, 15, 18]

# Let's start plotting
plt.scatter(x, y_1, label='Y values',
         color='blue', linestyle='dashed')
plt.scatter(x, y_2, label='Y2 values',
         color='violet', linestyle='dashed')
plt.xlabel('time')
plt.ylabel('depression')
plt.title('X vs Y')
plt.grid(alpha=.25)
plt.legend(loc='upper left') #loc is location
#plt.show()
#plt.savefig("XvsY.jpg") png pdf can also 
```

```python
# Some data for plotting
x = [0, 1, 2, 3, 4, 5, 6]
y_1 = [0, 2, 4, 6, 8, 10, 12]
y_2 = [0, 3, 6, 9, 12, 15, 18]
y_3 = [0, 3, 6, 9, 12, 15, 18]

# Let's start plotting
plt.scatter(x, y_1, label='Y values',
         color='blue', linestyle='dashed')
plt.scatter(x, y_2, label='Y2 values',
         color='violet', linestyle='dashed')

plt.xlabel('time')
plt.ylabel('depression')
plt.title('X vs Y')
plt.grid(alpha=.25) #transparency of grid
plt.legend(loc=[0,0.7]) #loc is location
#plt.show()
#plt.savefig("XvsY.jpg") png pdf can also 
```

```python
# Some data for plotting
x = [0, 1, 2, 3, 4, 5, 6]
y_1 = [0, 2, 4, 6, 8, 10, 12]
y_2 = [0, 3, 6, 9, 12, 15, 18]
y_3 = [0, 4, 8, 12, 16, 20, 24]
x_error, y_error = .1, 0.75

# Lets start plotting
plt.plot(x, y_1, label='Y1 values', color='red', linestyle='dashed',)
plt.plot(x, y_2, label='Y2 values', color='blue', )
plt.errorbar(x, y_3, xerr=x_error, yerr=y_error,
             label='Y3 with errors', color='green')
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=.25)
plt.legend(loc='upper left')
```

```python
x = np.array([0, 1, 2, 3, 4, 5, 6])

x2 = x**2                  # The math stuff
sin_x = np.sin(x)

plt.plot(x, x2, label='x^2',
         color='red', linestyle='dashed', )
plt.plot(x, sin_x, label='sin(x)',
         color='blue')
plt.legend()                                 
```

```python
x = np.array([0, 1, 2, 3, 4, 5, 6])

x=np.linspace(0,2*np.pi,1000) #to make it smooth, and add 1000 data intervals btw 0,1,2,3,4,5,6

x2 = x**2
sin_x = np.sin(x)

plt.plot(x, x2, label='x^2',color='red', linestyle='dashed')
plt.legend(loc='lower left')                                  # For y-axis 1

plt.twinx()                                                   # This creates a new y-axis 
                                                              # for the plots that comes after
plt.plot(x, sin_x, label='sin(x)',color='blue', )
plt.legend(loc='lower right')                                 # For y-axis 2
```
