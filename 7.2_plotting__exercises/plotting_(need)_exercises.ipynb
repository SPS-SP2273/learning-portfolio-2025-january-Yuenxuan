```python
import matplotlib.pyplot as plt
import numpy as np

# Data
data = [
    [39171, 24407, 55012, 58190, 48502, 49105, 56222, 64584, 55534, 26581],  # Manual mask area
    [33799, 25345, 33941, 31615, 30180, 32751, 38360, 43243, 37903, 18245],  # Algorithm mask area
    [28805, 17206, 32967, 30947, 29096, 32015, 36903, 40936, 36021, 16292]   # Overlap area
]

labels = ["Manual mask area (pixels)", "Algorithm mask area (pixels)", "Overlap area (pixels)"]

# Plot
plt.figure(figsize=(8, 6))
bp = plt.boxplot(data, patch_artist=True, showmeans=True, meanline=True, 
                 meanprops={"color": "red", "linewidth": 2})

# Change box colors
colors = ['yellow', 'orange', 'white']
for box, color in zip(bp['boxes'], colors):
    box.set(facecolor=color)

# Change median line color
for median in bp['medians']:
    median.set(color='blue', linewidth=2)

# Add median value labels
medians = [np.median(d) for d in data]
for i, median in enumerate(medians):
    plt.text(i + 1, median, f'{median:.0f}', horizontalalignment='center', 
             verticalalignment='bottom', fontname="Arial", fontsize=10, color='black')

# Customize plot
plt.xticks([1, 2, 3], labels, rotation=0, fontname="Arial")  # Set labels to be straight and font to Arial
plt.yticks(fontname="Arial")  # Set y-axis numbers font to Arial
plt.ylabel("Area (pixels)", fontname="Arial")
plt.ylim(0, 70000)  # Set y-axis range from 0 to 70000
plt.title("Comparison of Mask Areas", fontname="Arial")
plt.show()
```
